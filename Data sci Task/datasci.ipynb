Import Data

```python
import pandas as pd
df=pd.read_csv("drinkMenu.csv")
```

Data Preparation


```python
#check duplicates and remove them if founded 
df.duplicated().sum()
# there is no duplicates 
```

```python
#check null values and fill if founded 
df.isnull().sum()
```

```python
# fill null values with mode as it's a categorical field 
modecaff=df['Caffeine (mg)'].mode()
df['Caffeine (mg)'].fillna(int(modecaff),inplace=True)
```

```python
#check there is no null values
df.isnull().sum().sum()
```

```python
# remove unnecessary values 
df.drop(df.columns[[4,5,6,7,10,13,14,15,16]],axis=1,inplace=True)
```

```python
df.columns
```

Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
#Highest calorie drink
plt.barh(df['Beverage_category'],df['Calories'])
plt.title('Highest calorie drink')
plt.show()
```

```python
#Highest sugar drink
df.groupby(['Beverage_category']).sum().plot(kind='pie', y=' Sugars (g)', autopct='%1.0f%%')
plt.legend( bbox_to_anchor=(1.5,0.5), loc="center right", fontsize=10, 
           bbox_transform=plt.gcf().transFigure)
plt.title('Highest sugar drink')
plt.show()

```
